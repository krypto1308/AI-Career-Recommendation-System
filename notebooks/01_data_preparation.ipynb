# Job Market Intelligence Layer
# Role–Skill Knowledge Base Creation


```python
import pandas as pd
import os

BASE_DIR = os.path.abspath("..")
DATA_PATH = os.path.join(BASE_DIR, "data", "processed", "skills_processed.csv")

df = pd.read_csv(DATA_PATH)
df.head()
```

```python
df.info()
df.isnull().sum()
```

```python
role_skill_map = (
    df.groupby("Title")["Normalized_Skill"]
    .apply(list)
    .reset_index()
)

role_skill_map.head()
```

```python
role_skill_map["Role_Skills"] = role_skill_map["Normalized_Skill"].apply(
    lambda x: list(set(x))
)

role_skill_map = role_skill_map.drop(columns=["Normalized_Skill"])
role_skill_map.head()

```

```python
skill_demand = (
    df["Normalized_Skill"]
    .value_counts()
    .reset_index()
)

skill_demand.columns = ["Skill", "Demand_Count"]
skill_demand.head()
```

```python
skill_demand["Demand_Score"] = (
    skill_demand["Demand_Count"] / skill_demand["Demand_Count"].max()
)

skill_demand.head()
```

```python
exp_role_skill_map = (
    df.groupby(["ExperienceLevel", "Title"])["Normalized_Skill"]
    .apply(lambda x: list(set(x)))
    .reset_index()
)

exp_role_skill_map.head()
```

```python
all_skills = sorted(df["Normalized_Skill"].unique().tolist())
len(all_skills), all_skills[:10]
```

```python
role_skill_map.to_csv(
    os.path.join(BASE_DIR, "data", "processed", "role_skill_map.csv"),
    index=False
)

skill_demand.to_csv(
    os.path.join(BASE_DIR, "data", "processed", "skill_demand_scores.csv"),
    index=False
)

exp_role_skill_map.to_csv(
    os.path.join(BASE_DIR, "data", "processed", "exp_role_skill_map.csv"),
    index=False
)
```
